## 1. Intro to polygons in R

Now we'll go over some useful functions associated with drawing normal distributions.  First, a little intro of sequences and polygons:

```r
x = seq(-3,3, length=10)
print(x)
```

Now, lets try to understand a `polygon` function.  We'll use this to help us draw areas using the `plot_polygons.R` script, but let's look at a few `polygon` examples.

Let's make a triangle -- say the triangle goes from -3 to +3 in x & 0-1 in y:

```r
plot(NULL,xlim=c(-3,3),ylim=c(0,1)) # sets up axes
xvertices = c(-3, 0, 3)
yvertices = c(0, 1, 0)
polygon(xvertices, yvertices,col="red") # plots on top of previous plot
```

Let's try overplotting a little rectangle at x = (-1,1), y = (0.4,0.6):

```r
# set up empty axis
plot(NULL,xlim=c(-3,3),ylim=c(0,1)) # sets up axes

# red triangle
xvertices = c(-3, 0, 3)
yvertices = c(0, 1, 0)
polygon(xvertices, yvertices,col="red") # plots on top of previous plot

# blue rectangle
xvertices = c(-1, -1, 1, 1)
yvertices = c(0.4, 0.6, 0.6, 0.4)
polygon(xvertices, yvertices, col="blue")
```

Essentially, polygon just fills in between a list of verticies we give it.  We can use this to plot underneath our normal distributions.  This will help us get a "feel" for how much of the graph is related to our measurement of interest.

## 2. Plotting normal distributions

Now, let's build some tool's we will need to examine normal distributions.

(1) Let's plot them using "dnorm" moving onto normal distributions. First, let's start by plotting a normal distribution like last time:

```r
help(dnorm) # to see list of normal distribution functions
```

```r
x=seq(-3,3,length=200)
y=dnorm(x, mean=0, sd=1)
plot(x,y)
```

Center this around a different value:

```r
y = dnorm(x, mean=2, sd=1)
plot(x,y)
```

But, of course with a new mean comes a new sequence to be able to plot it:

```r
x=seq(-2,4,length=200)
plot(x, dnorm(x,mean=1,sd=1))
```

We can overplot a bunch of different normal distributions:

```r
# overall sequence
x = seq(-10,10,length=200)

# a few different normal distributions
y1 = dnorm(x,mean=0,sd=1)
y2 = dnorm(x,mean=0, sd=5)
y3 = dnorm(x,mean=0,sd=0.25)

# plot 
plot(x,y1, type="l", ylim=c(0,2))
par(new=TRUE) # overplot
plot(x,y2,type='l',col='red',ylim=c(0,2))
par(new=TRUE) # overplot
plot(x,y3,type='l',col='blue', ylim=c(0,2))
```

(2) 2nd in our toolset is calculating Z-scores: remember this is a measure of how "far off" a score is from the mean.

So first, as is always a good example, let's plot!

```r
x = seq(-6,6,length=200)
mean_dist = 1.0
sd_dist = 0.5
plot(x,dnorm(x,mean=mean_dist,sd=sd_dist),ylim=c(0,1.0))
```

Let's say I want the Zscore for x=2.5?  First of course, let's plot!

```r
plot(x,dnorm(x,mean=mean_dist,sd=sd_dist),ylim=c(0,1.0))
abline(v=2.5,col="red")
```

Recall: $Z_{score} = \frac{observation - mean}{SD}$

```r
Zscore = (2.5 - mean_dist)/sd_dist
print(Zscore)
```

Now, let's say I've got a 2nd distribution with mean = 0.5 and sd=2, is the Zscore at x=2.5 higher or lower than the first one? As always, lets start by plotting:

```r
# old plot
plot(x,dnorm(x,mean=mean_dist,sd=sd_dist),ylim=c(0,1.0))
abline(v=2.5,col="red")

mean_dist2 = 0.5
sd_dist2 = 2.0
par(new=TRUE)
plot(x,dnorm(x,mean=mean_dist2,sd=sd_dist2),col="blue",ylim=c(0,1.0))
```

So, by eye we can see that the red line falls at a higher y-value on the blue, 2nd distribution.  This tells us at x=2.5, we are closer to the mean on the 2nd distribution so we expect a lower Zscore, but lets find out!

```r
Zscore2 = (2.5-mean_dist2)/sd_dist2
print(Zscore2)
```

Indeed 1 < 3

Zscores allow us to in a sense "normalize" each normal distribution to allow for comparisions between them.  For example, if these distributions were measuring a test then a student that scored a 2.5 on both would have done better than the overall class distribution on the first test.

## 4. Percentiles in R

We'll use the `plot_polygons.R` script to draw some polygons on our normal distributions.  To do this, we have to source it:

```r
# assuming in same directory as this notebook file
source('plot_polygons.R')
```

If we recall the sum under the curve of our normal probability distribution has to == 1 given the laws of probability, we can associate every Zscore with a percentile. 

Let's go back to our toy example of something centered a 1:

```r
x = seq(-3,3,length=200)
y = dnorm(x,mean=0,sd=1)
plot(x,y,type='l')
```

What is the Zscore for x=0?

```r
Zscore = (0-0)/1 # = 0
```

What is the percentile?  i.e. what percent under the curve corresponds to a Zscore of 0 here?  First, let's plot with a polygon. Let's make sure we are only giving vertices up to Zscore = 0:

```r
# old plot
x = seq(-3,3,length=200)
y = dnorm(x,mean=0,sd=1)
plot(x,y,type='l')

# adding polygon
#   grab sequences from the z-score
x2 = seq(-3, Zscore,length=100)
y2 = dnorm(x2, mean=0, sd=1)
draw_polygon(x2,y2)
```

So, by eye we can see that its probably 1/2 or a percentile of 50%, but we can also get an exact number using "pnorm":

```r
print(pnorm(Zscore,mean=0,sd=1))
```

In fact, we can do this percentage-from-Zscore for any Zscore make a plot of the percentile associated with negative "X":

```r
plot(x,y,type='l')
Zscore_of_interest = 2
x2 = seq(-3, Zscore_of_interest,length=100)
y2 = dnorm(x2, mean=0, sd=1)
draw_polygon(x2,y2)

# print too
print(pnorm(Zscore_of_interest,mean=0,sd=1))
```

This is now showing results for the area < some Zscore.  What about if we want to know the area for > some Zscore?

```r
plot(x,y,type='l')

# draw polygon for area > Zscore
x2 = seq(Zscore_of_interest,3,length=100)
y2 = dnorm(x2,mean=0,sd=1)
draw_polygon(x2,y2)
```

Note: you can find this percentile using the `lower.tail=FALSE` flag:

```r
print(1-pnorm(Zscore_of_interest,mean=0,sd=1)) # this is equivalent to
print(pnorm(Zscore_of_interest,mean=0,sd=1,lower.tail=FALSE)) # this
```

I can also ask for the percentage of the curve between 2 numbers:

```r
x = seq(-3,3,length=200)
y = dnorm(x,mean=0,sd=1)
plot(x,y,type='l')
```

What percentage of the curve is between -1 and 1?  First, let's plot:

```r
x = seq(-3,3,length=200)
y = dnorm(x,mean=0,sd=1)
plot(x,y,type='l')

x2 = seq(-1,1,length=100)
y2 = dnorm(x2,mean=0,sd=1)
draw_polygon(x2,y2)
```

For the calculation of percentage, we have to subtract one from the other like so:

```r
print(pnorm(1.0,mean=0,sd=1)-pnorm(-1.0,mean=0,sd=1))
```
